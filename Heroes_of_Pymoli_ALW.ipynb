```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

Player Count

```python
# display total players

total_players = purchase_df["SN"].unique()
total_players = len(total_players)
print(total_players)

total_players_df = pd.DataFrame({"Total Players" : [total_players]})
total_players_df.head()
```

Purchasing Analysis (Total)

```python
#Run basic calculations to obtain number of unique items, average price, etc.
items_unique = purchase_df["Item ID"].unique()
items_unique = len(items_unique)
print(items_unique)

average_price = purchase_df["Price"].mean()
print(average_price)

total_purchases = len(purchase_df)
print(total_purchases)

total_rev = purchase_df["Price"].sum()
print(total_rev)

#Create a summary data frame to hold the results
purchase_analysis_df = pd.DataFrame({
    "Unique Items" : [items_unique],
    "Average Price" : [average_price],
    "Total Purchases" : [total_purchases],
    "Total Revenue" : [total_rev]
})

purchase_analysis_df["Average Price"] =  purchase_analysis_df["Average Price"].apply("${:,.2f}".format)
purchase_analysis_df["Total Purchases"] =  purchase_analysis_df["Total Purchases"].apply("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] =  purchase_analysis_df["Total Revenue"].apply("${:,.2f}".format)

purchase_analysis_df
#Optional: give the displayed data cleaner formatting


#Display the summary data frame
```

Gender Demographics
Percentage and Count of Male Players
Percentage and Count of Female Players
Percentage and Count of Other / Non-Disclosed

```python
gender_data = purchase_df.loc[:,["SN", "Gender"]]
gender_data = gender_data.drop_duplicates(subset="SN", keep="first", inplace=False)

gender_count = gender_data["Gender"].value_counts()

gender_percents = gender_count / total_players * 100

gender_df = pd.DataFrame({"Total Count" : gender_count, "Percentage of Players" : gender_percents})
gender_df["Percentage of Players"] =  gender_df["Percentage of Players"].apply("{:,.0f}%".format)
gender_df.head()
```

 Purchasing Analysis (Gender)
 
Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

Create a summary data frame to hold the results

Optional: give the displayed data cleaner formatting

Display the summary data frame

```python
#should it still be de-duped by SN or just take entire df
gender_data = purchase_df.groupby(["Gender"])
print(type(gender_data))

purchase_count = gender_data["Item Name"].count()
purchase_average = gender_data["Price"].mean()
purchase_value = gender_data["Price"].sum()

per_person = purchase_df.drop_duplicates(subset="SN", keep="first", inplace=False)
grouped_per_person = per_person.groupby(["Gender"])

average_per_person = (gender_data["Price"].sum() / grouped_per_person["SN"].count())

purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": purchase_average,
                              "Total Purchase Value": purchase_value,
                              "Avg Total Purchase per Person": average_per_person})

purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].apply("${:,.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].apply("${:,.2f}".format)
purchase_analysis["Avg Total Purchase per Person"] = purchase_analysis["Avg Total Purchase per Person"].apply("${:,.2f}".format)

purchase_analysis
```

Age Demographics
 
Establish bins for ages

Categorize the existing players using the age bins. Hint: use pd.cut()

Calculate the numbers and percentages by age group

Create a summary data frame to hold the results

Display Age Demographics Table

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["less than 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40 and up"]
```

```python
purchase_df["Age Demographics"]= pd.cut(purchase_df["Age"], age_bins, labels=group_names)
age_demo = purchase_df.drop_duplicates(subset = "SN", keep="first", inplace=False)

#google searched - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.to_frame.html
age_demo_df = (age_demo["Age Demographics"].value_counts()).to_frame()

age_demo_df = age_demo_df.rename(columns={"Age Demographics" : "Total Count"})
age_demo_df.head()

age_demo_df["Percentage of Players"] = age_demo_df["Total Count"] / [total_players] * 100

age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].apply("{:,.2f}%".format)

age_demo_df
```

Purchasing Analysis (Age)
Bin the purchase_data data frame by age

Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below

Create a summary data frame to hold the results

```python
#age_demo is the binned data frame by age

purchase_df.head()

purchase_data = purchase_df.loc[:,["Age Demographics", "Price"]]

purchase_analysis = purchase_data["Age Demographics"].value_counts().to_frame()
purchase_analysis = purchase_analysis.rename(columns={"Age Demographics" : "Purchase Count"})


avg_purchase_price = purchase_df["Price"].groupby(purchase_df["Age Demographics"]).mean()
purchase_analysis["Average Purchase Price"] = avg_purchase_price


total_purchase_value = purchase_df["Price"].groupby(purchase_df["Age Demographics"]).sum()
purchase_analysis["Total Purchase Value"] = total_purchase_value

purchase_analysis["Average Total Price Per Person"] = purchase_analysis["Total Purchase Value"] / purchase_analysis["Purchase Count"]

purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].apply("${:,.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].apply("${:,.2f}".format)
purchase_analysis["Average Total Price Per Person"]= purchase_analysis["Average Total Price Per Person"].apply("${:,.2f}".format)

purchase_analysis

```

```python
#define top spender 
top_spenders = purchase_df["Item ID"].groupby(purchase_df["SN"]).count()
type(top_spenders)

#make it a df
top_spenders_df= pd.DataFrame(top_spenders)
top_spenders_df.columns = ["Purchase Count"]

#get the summary data(avg price & total purchase value by top spender (sn))
top_spenders_df["Average Purchase Price"] = purchase_df["Price"].groupby(purchase_df["SN"]).mean()
top_spenders_df["Total Purchase Value"] = purchase_df["Price"].groupby(purchase_df["SN"]).sum()


top_spenders_df.sort_values(by=["Total Purchase Value"], ascending=False, inplace=True)

top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)

top_spenders_df.head()
```
